### Methodology:
By first invesigation, the dataset include information of 77 Customers, 579 SKU in 10 SKU Groups. It is not realistic to analyze the dataset as a whole, or treat all informatoion as the same, because the statistic is made in 3 years (2017-2019). The Strategy to analyze the information should be:
 - Analyze by year separately, as the information of 2019 would contribute more than 2018 and 2017
 - Categorize Customers by 3 levels base on their Order Value (top 25%, from 75%-25%, and the rest)
 - Categorize SKU and SKU Group by 3 levels base on their Order Value (top 25%, from 75%-25%, and the rest)
 - Besides, preparing for the delivery base on distance from DC to top Retail Store should be considered as well, but we face the limitation of location information in the dataset.

#### For Customer analyze, we work with Customer Codes that >0

```python
df_EA_neg=df.loc[df['Extended Amount']<0]
df=df.loc[df['Extended Amount']>=0]
```

#### Create more columns regarding Year and Month of the Order, and lead time between order and delivery

```python
# Create some more columns to categorize Orders by Month and Year
df['Request_year']=pd.DatetimeIndex(df['Request Date']).year
df['Request_month']=pd.DatetimeIndex(df['Request Date']).month
df['Request_to_Shipment']=pd.to_datetime(df['Actual Ship'])-pd.to_datetime(df['Request Date'])
df['Order_to_Request']=pd.to_datetime(df['Request Date'])-pd.to_datetime(df['Order Date'])
```

###  Fill missing Value in Unit Price
We can see that the std of Unit Price is high, so missing value should be filled by median of each group (grouped by year, month and SKU)

```python
df['Unit Price'].replace(0,np.nan, inplace=True)
df['Unit Price']=df[['Request_year','Request_month','2nd Item Number','Unit Price']].groupby(['Request_year','Request_month','2nd Item Number']).transform(lambda x: x.fillna(x.max()))
```

```python
df_2017=df.loc[df['Request_year']==2017]
df_2018=df.loc[df['Request_year']==2018]
df_2019=df.loc[df['Request_year']==2019]
```

#### We start to Categorize Customer base on Value of Order:

```python
Customer_2019=df_2019[['Sold To','Extended Amount']].groupby(['Sold To'], as_index=False).sum().sort_values('Extended Amount',ascending=False)
Customer_2018=df_2018[['Sold To','Extended Amount']].groupby(['Sold To'], as_index=False).sum().sort_values('Extended Amount',ascending=False)
Customer_2017=df_2017[['Sold To','Extended Amount']].groupby(['Sold To'], as_index=False).sum().sort_values('Extended Amount',ascending=False)
SKU_2019=df_2019[['2nd Item Number','Extended Amount']].groupby(['2nd Item Number'], as_index=False).sum().sort_values('Extended Amount',ascending=False)
SKU_2018=df_2018[['2nd Item Number','Extended Amount']].groupby(['2nd Item Number'], as_index=False).sum().sort_values('Extended Amount',ascending=False)
SKU_2017=df_2017[['2nd Item Number','Extended Amount']].groupby(['2nd Item Number'], as_index=False).sum().sort_values('Extended Amount',ascending=False)
```

### Customer Categorize:
Using  Principal, we define Customer that contribute 80% of Extended Amount to Class_A, the remaining in group B

```python
P_A=df_2019['Extended Amount'].sum()*0.8
P_B=df_2019['Extended Amount'].sum()*0.90
total=0
Custome_A=[]
Custome_B=[]
Custome_C=[]
Customer_2019['Customer_2019_Cat']='NaN'
for index, row in Customer_2019.iterrows():
    total+=row['Extended Amount']
    if total<P_A:
        Custome_A.append(row['Sold To'])
    elif total>P_A and total<P_B:
        Custome_B.append(row['Sold To'])
    else:
        Custome_C.append(row['Sold To'])
            
for index, row in Customer_2019.iterrows():
    if Customer_2019.iloc[index][0] in Custome_A:
        Customer_2019['Customer_2019_Cat'].iloc[index]="A"
    elif Customer_2019.iloc[index][0] in Custome_B:
        Customer_2019['Customer_2019_Cat'].iloc[index]="B"
    else:
        Customer_2019['Customer_2019_Cat'].iloc[index]="C"
Customer_2019
```

```python
P_A=df_2018['Extended Amount'].sum()*0.8
P_B=df_2018['Extended Amount'].sum()*0.90
total=0
Custome_A=[]
Custome_B=[]
Custome_C=[]
Customer_2018['Customer_2018_Cat']='NaN'
for index, row in Customer_2018.iterrows():
    total+=row['Extended Amount']
    if total<P_A:
        Custome_A.append(row['Sold To'])
    elif total>P_A and total<P_B:
        Custome_B.append(row['Sold To'])
    else:
        Custome_C.append(row['Sold To'])
            
for index, row in Customer_2018.iterrows():
    if Customer_2018.iloc[index][0] in Custome_A:
        Customer_2018['Customer_2018_Cat'].iloc[index]="A"
    elif Customer_2018.iloc[index][0] in Custome_B:
        Customer_2018['Customer_2018_Cat'].iloc[index]="B"
    else:
        Customer_2018['Customer_2018_Cat'].iloc[index]="C"
Customer_2018
```

### SKU Categorize:

```python

A=SKU_2019.describe(include='all').loc['75%'][1]
B=SKU_2019.describe(include='all').loc['50%'][1]
conditions=[SKU_2019['Extended Amount']>=A,(SKU_2019['Extended Amount']>=B)&(SKU_2019['Extended Amount']<A),SKU_2019['Extended Amount']<B]
values=['A','B','C']
SKU_2019['SKU_2019_Cat'] = np.select(conditions, values)
```
